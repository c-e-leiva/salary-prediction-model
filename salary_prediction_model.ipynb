## üìö Importaci√≥n de librer√≠as

Se cargan las bibliotecas esenciales para manipulaci√≥n de datos, codificaci√≥n de variables, escalado, divisi√≥n de conjuntos, entrenamiento y evaluaci√≥n de modelos de machine learning, junto con herramientas para interacci√≥n en el notebook.


```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import ipywidgets as widgets
from IPython.display import display, clear_output
```

## üßπüî¢ Limpieza y codificaci√≥n de datos

Se eliminan los valores nulos y se codifican las variables categ√≥ricas con LabelEncoder, que transforma textos en n√∫meros para que puedan ser interpretados por los algoritmos de machine learning.

```python
# Limpiar datos
df = df.dropna()  # Eliminamos filas con datos faltantes por seguridad

# Variables categ√≥ricas y num√©ricas
cat_cols = ['Job Title', 'Company', 'Location', 'Experience Level', 'Industry', 'Required Skills']
target_col = 'Salary'

# Codificamos las variables categ√≥ricas con LabelEncoder
#for col in cat_cols:
#    df[col] = LabelEncoder().fit_transform(df[col].astype(str))
```

## üí∞üìä Clasificaci√≥n del salario en categor√≠as

Se crean tres categor√≠as salariales (Inferior, Promedio y Superior) seg√∫n rangos definidos, y se asignan a una nueva columna llamada `Salary Category`.



```python
# Clasificar los salarios en 3 categor√≠as: Inferior, Promedio y Superior
def clasificar_salario(salario):
    if salario < 68000:
        return 'Salario Inferior'
    elif 68000 <= salario <= 123000:
        return 'Salario Promedio'
    else:
        return 'Salario Superior'

df['Salary Category'] = df[target_col].apply(clasificar_salario)

# Clasificamos los salarios en tres categor√≠as utilizando los percentiles del resumen estad√≠stico:
# - Menor al percentil 25 (68000): 'Salario Inferior'
# - Entre el percentil 25 y el 75 (68000 a 123000): 'Salario Promedio'
# - Mayor al percentil 75 (123000): 'Salario Superior'
# Esto nos permite segmentar los sueldos seg√∫n su distribuci√≥n en el conjunto de datos.
```

## üéØ Definici√≥n de variables para el modelo

Se redefine la variable objetivo como `Salary Category` y se separan las variables independientes (features) y la variable dependiente (target) para el entrenamiento del modelo.

```python
# Ahora cambiamos el objetivo de 'Salary' a 'Salary Category'
target_col = 'Salary Category'

# Variables independientes (features) y la variable dependiente (target)
X = df.drop(columns=[target_col])
y = df[target_col]
```

## üìä Codificaci√≥n y estandarizaci√≥n de variables

Se codifican nuevamente las variables categ√≥ricas del conjunto `X` con `LabelEncoder`, y se estandarizan las variables num√©ricas usando `StandardScaler` para mejorar el rendimiento de los modelos.


```python
# Codificamos las variables categ√≥ricas
for col in cat_cols:
    X[col] = LabelEncoder().fit_transform(X[col].astype(str))

# Estandarizamos solo las variables num√©ricas (features)
num_cols = X.select_dtypes(include=[np.number]).columns.tolist()
scaler = StandardScaler()
X[num_cols] = scaler.fit_transform(X[num_cols])
```

## üîÄ A√±adir ruido a los datos

Se define una funci√≥n que agrega ruido aleatorio a las variables num√©ricas, con el fin de aumentar la complejidad del modelo y evitar sobreajuste. La semilla garantiza reproducibilidad del experimento.


```python
# Funci√≥n para agregar ruido a los datos (m√°s ruidoso)
def add_noise(data, noise_level=0.1):
    noisy_data = data.copy()
    for col in noisy_data.select_dtypes(include=['float64', 'int64']).columns:
        noise = np.random.normal(0, noise_level, noisy_data[col].shape)
        noisy_data[col] = noisy_data[col] + noise
    return noisy_data

np.random.seed(42)  # Semilla fija para ruido reproducible
# Agregar ruido a las caracter√≠sticas
X_noisy = add_noise(X, noise_level=0.1)  # Ruido aumentado
```

## üß™ Divisi√≥n del dataset en entrenamiento y prueba
Se divide el conjunto de datos en entrenamiento (80%) y prueba (20%) usando `train_test_split`, para evaluar el rendimiento de los modelos con datos no vistos.



```python
# Divisi√≥n en train y test
X_train, X_test, y_train, y_test = train_test_split(X_noisy, y, test_size=0.2, random_state=42)
```

## ü§ñ Entrenamiento y optimizaci√≥n de modelos de clasificaci√≥n  
Se entrenan tres modelos de clasificaci√≥n utilizando los datos de entrenamiento y se optimizan sus hiperpar√°metros con `GridSearchCV` para mejorar su desempe√±o:

- **Regresi√≥n Log√≠stica**  
- **√Årbol de Decisi√≥n**  
- **Random Forest**  

GridSearchCV realiza una b√∫squeda exhaustiva en una rejilla de par√°metros, usando validaci√≥n cruzada para encontrar la mejor combinaci√≥n de hiperpar√°metros y evitar sobreajuste.


```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV

# === Logistic Regression
param_grid_lr = {
    'C': [0.01, 0.1],
    'solver': ['liblinear', 'lbfgs']
}

grid_lr = GridSearchCV(LogisticRegression(max_iter=500, random_state=42), param_grid_lr, cv=2)
grid_lr.fit(X_train, y_train)
y_pred_lr = grid_lr.predict(X_test)
print("\nLogistic Regression best params:", grid_lr.best_params_)
print("Accuracy Logistic Regression:", grid_lr.score(X_test, y_test))


# === Decision Tree
param_grid_dt = {
    'max_depth': [2, 3],
    'min_samples_split': [4, 5]
}

grid_dt = GridSearchCV(DecisionTreeClassifier(random_state=42), param_grid_dt, cv=2)
grid_dt.fit(X_train, y_train)
y_pred_dt = grid_dt.predict(X_test)
print("\nDecision Tree best params:", grid_dt.best_params_)
print("Accuracy Decision Tree:", grid_dt.score(X_test, y_test))


# === Random Forest (rangos amplios para mejorar) ===
param_grid_rf = {
    'n_estimators': [40, 50],     # m√°s √°rboles para mejorar estabilidad
    'max_depth': [5, 10, None],    # mayor profundidad para mayor complejidad
    'min_samples_split': [2, 3]          # pocos splits para flexibilidad
}

grid_rf = GridSearchCV(RandomForestClassifier(random_state=42), param_grid_rf, cv=5)
grid_rf.fit(X_train, y_train)
y_pred_rf = grid_rf.predict(X_test)
print("\nRandom Forest best params:", grid_rf.best_params_)
print("Accuracy Random Forest:", grid_rf.score(X_test, y_test))
```

## üìä Evaluaci√≥n del rendimiento de los modelos  
Se define una funci√≥n para evaluar y mostrar las m√©tricas clave de desempe√±o de cada modelo entrenado, tales como:

- Precisi√≥n (accuracy)  
- Matriz de confusi√≥n  
- Reporte de clasificaci√≥n (precision, recall, f1-score)  

Estas m√©tricas permiten comparar la efectividad y calidad de las predicciones de cada modelo.




```python
def print_metrics(y_true, y_pred, model_name):
    accuracy = accuracy_score(y_true, y_pred)
    print(f"{model_name} ")
    print(f"Accuracy: {accuracy:.4f}")
    print(confusion_matrix(y_true, y_pred))
    print(classification_report(y_true, y_pred))
    print("\n")

# Imprimir las m√©tricas para cada modelo
print_metrics(y_test, y_pred_lr, "Regresi√≥n Log√≠stica")
print_metrics(y_test, y_pred_dt, "√Årbol de Decisi√≥n")
print_metrics(y_test, y_pred_rf, "Random Forest")
```

## üìà An√°lisis de la Matriz de Confusi√≥n  
Se visualizan las matrices de confusi√≥n de los modelos **Regresi√≥n Log√≠stica**, **√Årbol de Decisi√≥n** y **Random Forest** usando Plotly, para evaluar su desempe√±o en la clasificaci√≥n de las tres categor√≠as salariales:  

- Salario Inferior  
- Salario Promedio  
- Salario Superior  

Esta visualizaci√≥n interactiva facilita la interpretaci√≥n de los aciertos y errores de cada modelo.

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import pandas as pd
from sklearn.metrics import confusion_matrix

# Etiquetas de clase
labels = ["Salario Inferior", "Salario Promedio", "Salario Superior"]

# Crear la figura con 3 subgr√°ficos horizontales
fig = make_subplots(
    rows=1, cols=3,
    subplot_titles=("Regresi√≥n Log√≠stica", "√Årbol de Decisi√≥n", "Random Forest"),
    shared_yaxes=True
)

# Funci√≥n para agregar cada matriz a los subplots
def add_confusion_to_subplot(fig, y_true, y_pred, col, scale='Reds'):
    cm = confusion_matrix(y_true, y_pred)
    fig.add_trace(
        go.Heatmap(
            z=cm,
            x=labels, # Predicci√≥n del modelo
            y=labels, # Clase real
            colorscale=scale,
            colorbar=dict(title="Count") if col == 3 else None,
            showscale=True if col == 3 else False,
            hovertemplate="Clase real: %{y}<br>Predicci√≥n del modelo: %{x}<br>Count: %{z}<extra></extra>"

        ),
        row=1, col=col
    )

# Agregar los 3 modelos
add_confusion_to_subplot(fig, y_test, y_pred_lr, col=1, scale='Blues')
add_confusion_to_subplot(fig, y_test, y_pred_dt, col=2, scale='Blues')
add_confusion_to_subplot(fig, y_test, y_pred_rf, col=3, scale='Blues')

# Ajustes de layout
fig.update_layout(
    title_text="Matrices de Confusi√≥n Comparadas por Modelo",
    height=500,
    width=1300
)

# Mostrar figura
fig.show()
```

## üìä Comparaci√≥n de Precisi√≥n entre Modelos  
Se comparan los modelos `'Regresi√≥n Log√≠stica'`, `'√Årbol de Decisi√≥n'` y `'Random Forest'` en funci√≥n de su precisi√≥n, permitiendo identificar cu√°l ofrece el mejor rendimiento para la clasificaci√≥n de categor√≠as salariales.  
La comparaci√≥n se realiza mediante un gr√°fico de barras horizontales generado con Plotly para una visualizaci√≥n clara y efectiva.

```python
import plotly.graph_objects as go
from sklearn.metrics import accuracy_score, classification_report

# Definir una funci√≥n para obtener precisi√≥n y otras m√©tricas importantes
def get_metrics(y_true, y_pred):
    accuracy = accuracy_score(y_true, y_pred)
    report = classification_report(y_true, y_pred, output_dict=True)
    return accuracy, report['accuracy'], report['macro avg']['f1-score']

# Calcular m√©tricas para cada modelo
lr_accuracy, lr_accuracy_detail, lr_f1 = get_metrics(y_test, y_pred_lr)
dt_accuracy, dt_accuracy_detail, dt_f1 = get_metrics(y_test, y_pred_dt)
rf_accuracy, rf_accuracy_detail, rf_f1 = get_metrics(y_test, y_pred_rf)

# Crear un diccionario con los resultados de precisi√≥n
models = {
    'Regresi√≥n Log√≠stica': lr_accuracy,
    '√Årbol de Decisi√≥n': dt_accuracy,
    'Random Forest': rf_accuracy
}

# Colores para cada modelo
colors = {
    'Regresi√≥n Log√≠stica': '#1f77b4',  # azul
    '√Årbol de Decisi√≥n': '#ff7f0e',    # naranja
    'Random Forest': '#2ca02c'          # verde
}

# Ordenar los modelos por precisi√≥n (de mayor a menor)
sorted_models = sorted(models.items(), key=lambda x: x[1], reverse=True)

# Preparar los datos ordenados
model_names = [model[0] for model in sorted_models]
accuracies = [model[1] for model in sorted_models]

# Crear gr√°fico de barras horizontal
fig = go.Figure(data=[
    go.Bar(
        y=[model_names[i]],
        x=[accuracies[i]],
        orientation='h',  # barras horizontales
        text=[f"{accuracies[i]:.4f}"],  # texto con precisi√≥n
        textposition='auto',
        textfont=dict(size=16),
        marker=dict(color=colors[model_names[i]])
    ) for i in range(len(model_names))
])

# Mejoras de estilo
fig.update_layout(
    title='Comparaci√≥n de Modelos - Precisi√≥n',
    xaxis_title='Precisi√≥n',
    yaxis_title='Modelo',
    yaxis=dict(autorange='reversed'),  # para mostrar la barra de mayor a menor de arriba a abajo
    width=1300,
    height=500,
    showlegend=False,
)

fig.show()
```

## üìà Reporte de Impacto Econ√≥mico de Modelos de Clasificaci√≥n üí∞

Este m√≥dulo permite generar un reporte comparativo del impacto econ√≥mico para los modelos de clasificaci√≥n seleccionados:

**Random Forest**, **√Årbol de Decisi√≥n** y **Regresi√≥n Log√≠stica**.

El reporte incluye una tabla detallada que muestra, por cada categor√≠a salarial (Salario Inferior, Promedio y Superior):
- La cantidad de aciertos (verdaderos positivos)
- Errores de tipo falso positivo y falso negativo
- Beneficios obtenidos por acierto
- Penalizaciones por errores
- Impacto econ√≥mico neto

La interfaz cuenta con dos botones:
- **Generar Reporte:** despliega la tabla con los datos anteriores.
- **Mostrar Gr√°fico:** presenta visualizaciones de barras que ilustran:
  - Impacto econ√≥mico por categor√≠a salarial en USD
  - Penalizaci√≥n total, aciertos totales e impacto neto total para el modelo seleccionado

Estas herramientas facilitan una comparaci√≥n clara y pr√°ctica del desempe√±o econ√≥mico de cada modelo.

```python
import ipywidgets as widgets
from IPython.display import display, clear_output, HTML
import pandas as pd
import numpy as np
from sklearn.metrics import confusion_matrix
import altair as alt

# Diccionario de modelos y sus predicciones
modelos_dict = {
    "Random Forest": {"y_pred": y_pred_rf, "modelo": grid_rf},
    "√Årbol de Decisi√≥n": {"y_pred": y_pred_dt, "modelo": grid_dt},
    "Regresi√≥n Log√≠stica": {"y_pred": y_pred_lr, "modelo": grid_lr},
}

# Widgets
selector_modelo = widgets.RadioButtons(
    options=list(modelos_dict.keys()),
    description='Modelo:',
    style={'description_width': 'initial'}
)
boton_generar = widgets.Button(description="Generar Reporte", button_style="success")
boton_grafico = widgets.Button(description="Mostrar Gr√°fico", button_style="info")
salida = widgets.Output()
salida_grafico = widgets.Output()

df_impacto_global = None  # Variable global
modelo_actual = None      # Para guardar nombre del modelo seleccionado

# Tabla con impacto econ√≥mico
def tabla_impacto_economico(y_true, y_pred):
    valores = {
        'Salario Inferior':    {'beneficio': 50,   'costo_fp': 200, 'costo_fn': 100},
        'Salario Promedio':    {'beneficio': 80,   'costo_fp': 250, 'costo_fn': 150},
        'Salario Superior':    {'beneficio': 100,  'costo_fp': 500, 'costo_fn': 200},
    }
    etiquetas = ['Salario Inferior', 'Salario Promedio', 'Salario Superior']
    cm = confusion_matrix(y_true, y_pred, labels=etiquetas)

    datos = []
    total_beneficio = 0
    total_penalizacion = 0

    for i, clase in enumerate(etiquetas):
        vp = cm[i, i]
        fn = cm[i, :].sum() - vp
        fp = cm[:, i].sum() - vp

        beneficio = vp * valores[clase]['beneficio']
        penalizacion = (fp * valores[clase]['costo_fp']) + (fn * valores[clase]['costo_fn'])
        impacto_neto = beneficio - penalizacion

        total_beneficio += beneficio
        total_penalizacion += penalizacion

        datos.append({
            "Clase": clase,
            "Aciertos (VP)": vp,
            "Errores FP": fp,
            "Errores FN": fn,
            "Beneficio por Acierto": beneficio,
            "Penalizaci√≥n": penalizacion,
            "Impacto Neto": impacto_neto
        })

    total_neto = total_beneficio - total_penalizacion
    datos.append({
        "Clase": "üí∞ TOTAL üí∞",
        "Aciertos (VP)": "",
        "Errores FP": "",
        "Errores FN": "",
        "Beneficio por Acierto": total_beneficio,
        "Penalizaci√≥n": total_penalizacion,
        "Impacto Neto": total_neto
    })

    df = pd.DataFrame(datos)

    # Formatear columnas monetarias
    df_tabla = df.copy()
    for col in ['Beneficio por Acierto', 'Penalizaci√≥n', 'Impacto Neto']:
        df_tabla[col] = df_tabla[col].apply(lambda x: f"USD $ {x:,.0f}" if pd.notna(x) and x != "" else x)

    display(HTML(df_tabla.to_html(index=False)))
    return df

# Gr√°fico con Altair
def graficar_impacto(df, nombre_modelo):
    df_grafico = df[df["Clase"] != "üí∞ TOTAL üí∞"].copy()

    orden_metricas = ['Penalizaci√≥n', 'Beneficio por Acierto', 'Impacto Neto']
    df_melt = df_grafico.melt(
        id_vars='Clase',
        value_vars=orden_metricas,
        var_name='M√©trica',
        value_name='Valor'
    )

    colores = alt.Scale(
        domain=orden_metricas,
        range=['#e74c3c', '#27ae60', '#2980b9']
    )

    chart_clases = alt.Chart(df_melt).mark_bar().encode(
        x=alt.X('Clase:N', title='Clase', axis=alt.Axis(labelAngle=30)),
        y=alt.Y('Valor:Q', title='Monto - USD'),
        color=alt.Color('M√©trica:N', scale=colores, title='M√©trica'),
        xOffset='M√©trica:N',
        tooltip=['Clase', 'M√©trica', alt.Tooltip('Valor', format=',')]
    ).properties(
        width=600,
        height=400,
        title=f'Impacto Econ√≥mico por Clase del Modelo: {nombre_modelo}'
    )

    # Texto encima de las barras
    text = alt.Chart(df_melt).mark_text(
        align='center',
        baseline='bottom',
        dy=-5,
        fontSize=12
    ).encode(
        x=alt.X('Clase:N', axis=alt.Axis(labelAngle=30)),
        y='Valor:Q',
        text=alt.Text('Valor:Q', format=','),
        detail='M√©trica:N',
        xOffset='M√©trica:N'
    )

    chart_clases = (chart_clases + text)

    # Gr√°fico resumen total
    df_total = df[df["Clase"] == "üí∞ TOTAL üí∞"].copy()
    resumen_total = pd.DataFrame({
        "M√©trica": orden_metricas,
        "Valor": [
            df_total["Penalizaci√≥n"].values[0],
            df_total["Beneficio por Acierto"].values[0],
            df_total["Impacto Neto"].values[0]
        ]
    })

    chart_total = alt.Chart(resumen_total).mark_bar().encode(
        x=alt.X('M√©trica:N', title='M√©trica', axis=alt.Axis(labelAngle=0)),
        y=alt.Y('Valor:Q', title='Monto Total - USD'),
        color=alt.Color('M√©trica:N', legend=None, scale=colores),
        tooltip=[alt.Tooltip('M√©trica'), alt.Tooltip('Valor', format=',')]
    ).properties(
        width=600,
        height=400,
        title=f'üíº Resumen Econ√≥mico Total del Modelo: {nombre_modelo}'
    )

    # Texto en barras del resumen total
    text_total = alt.Chart(resumen_total).mark_text(
        align='center',
        baseline='bottom',
        dy=-5,
        fontSize=12
    ).encode(
        x='M√©trica:N',
        y='Valor:Q',
        text=alt.Text('Valor:Q', format=',')
    )

    chart_total = (chart_total + text_total)

    combined_chart = alt.hconcat(chart_clases, chart_total).resolve_scale(y='independent')

    # Guardar los gr√°ficos como variables globales para futura exportaci√≥n
    global altair_chart_clases, altair_chart_total
    altair_chart_clases = chart_clases
    altair_chart_total = chart_total

    with salida_grafico:
        clear_output()
        display(combined_chart)

# Acci√≥n del bot√≥n Generar Reporte
def generar_reporte(b):
    global df_impacto_global, modelo_actual
    with salida:
        clear_output()
        modelo_actual = selector_modelo.value
        y_pred = modelos_dict[modelo_actual]["y_pred"]
        df_impacto_global = tabla_impacto_economico(y_test, y_pred)
    # Limpiar tambi√©n la salida del gr√°fico al generar un reporte nuevo
    with salida_grafico:
        clear_output()

# Acci√≥n del bot√≥n Mostrar Gr√°fico
def mostrar_grafico(b):
    if df_impacto_global is not None and modelo_actual is not None:
        graficar_impacto(df_impacto_global, modelo_actual)
    else:
        with salida_grafico:
            clear_output()
            print("Primero genera el reporte.")

# Conectar botones
boton_generar.on_click(generar_reporte)
boton_grafico.on_click(mostrar_grafico)

# Mostrar interfaz
display(selector_modelo, widgets.HBox([boton_generar, boton_grafico]), salida, salida_grafico)
```

## üíæ Exportar Reporte Econ√≥mico en PDF  
Este bloque genera un bot√≥n llamado **"ExportarReporte.PDF"** que permite descargar un reporte completo del impacto econ√≥mico del modelo seleccionado.  

El reporte incluye:  
- Un t√≠tulo descriptivo del an√°lisis econ√≥mico.  
- La tabla de impacto econ√≥mico con detalles por clase (aciertos, errores, beneficios y penalizaciones).  
- Los gr√°ficos de barras que muestran el impacto econ√≥mico por categor√≠a salarial (Salario Inferior, Promedio y Superior).  
- Un resumen econ√≥mico total del modelo seleccionado con visualizaciones comparativas.  

Esta funcionalidad facilita compartir y analizar los resultados econ√≥micos del modelo de manera clara y profesional.

```python
import ipywidgets as widgets
from IPython.display import display, clear_output, FileLink, IFrame
import os
import re

# funci√≥n exportadora
from reportlab.lib.pagesizes import letter, landscape
from reportlab.platypus import SimpleDocTemplate, Table, TableStyle, Paragraph, Image, Spacer
from reportlab.lib import colors
from reportlab.lib.styles import getSampleStyleSheet

def exportar_reporte_pdf_reportlab(nombre_modelo, df_tabla, chart1, chart2, ruta_archivo):
    chart1_path = "chart1.png"
    chart2_path = "chart2.png"
    chart1.save(chart1_path, scale_factor=2)
    chart2.save(chart2_path, scale_factor=2)

    doc = SimpleDocTemplate(ruta_archivo, pagesize=landscape(letter),
                            rightMargin=30, leftMargin=30, topMargin=30, bottomMargin=30)
    elements = []

    styles = getSampleStyleSheet()
    style_title = styles['Title']
    style_normal = styles['Normal']

    elements.append(Paragraph(f"Reporte del Modelo: {nombre_modelo}", style_title))
    elements.append(Spacer(1, 12))

    data = [df_tabla.columns.tolist()]

    def formatear_valor(valor, col):
        if col in ['Beneficio por Acierto', 'Penalizaci√≥n', 'Impacto Neto']:
            try:
                num = float(str(valor).replace(',', '').replace('$', ''))
                return f"${num:,.0f}"
            except:
                return str(valor)
        else:
            return str(valor)

    for _, row in df_tabla.iterrows():
        fila = [formatear_valor(row[col], col) for col in df_tabla.columns]
        data.append(fila)

    num_cols = len(df_tabla.columns)
    page_width = landscape(letter)[0] - 60
    col_width = page_width / num_cols
    col_widths = [col_width] * num_cols

    table = Table(data, colWidths=col_widths, hAlign='CENTER')

    style = TableStyle([
        ('BACKGROUND', (0,0), (-1,0), colors.HexColor('#d3d3d3')),
        ('TEXTCOLOR', (0,0), (-1,0), colors.black),
        ('ALIGN', (0,0), (0,-1), 'LEFT'),
        ('ALIGN', (1,0), (-1,-1), 'RIGHT'),
        ('FONTNAME', (0,0), (-1,0), 'Helvetica-Bold'),
        ('FONTSIZE', (0,0), (-1,-1), 8),
        ('BOTTOMPADDING', (0,0), (-1,0), 8),
        ('GRID', (0,0), (-1,-1), 0.5, colors.black),
    ])

    table.setStyle(style)

    elements.append(table)
    elements.append(Spacer(1, 20))

    img1 = Image(chart1_path, width=450, height=250)
    img2 = Image(chart2_path, width=450, height=250)
    elements.append(img1)
    elements.append(Spacer(1, 12))
    elements.append(img2)

    doc.build(elements)

    os.remove(chart1_path)
    os.remove(chart2_path)

    print(f"‚úÖ PDF generado con reportlab: {ruta_archivo}")


# --- Variables de ejemplo ---
# modelo_actual = "Modelo X"
# df_impacto_global = dataframe
# altair_chart_clases = grafico1
# altair_chart_total = grafico2
ruta_pdf = f"reporte_modelo_impacto_Economico.pdf"

boton_exportar = widgets.Button(description="ExportarReporte.PDF", button_style='success')
output = widgets.Output()

def on_exportar_clicked(b):
    with output:
        clear_output()
        exportar_reporte_pdf_reportlab(modelo_actual, df_impacto_global, altair_chart_clases, altair_chart_total, ruta_pdf)
        if os.path.exists(ruta_pdf):
            display(widgets.HTML(f'<b>Archivo generado:</b> {ruta_pdf}'))
            display(FileLink(ruta_pdf))
            display(IFrame(src=ruta_pdf, width=1000, height=400))

boton_exportar.on_click(on_exportar_clicked)
display(boton_exportar, output)
```
