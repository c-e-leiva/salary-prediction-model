## 📚 Importación de librerías

Se cargan las bibliotecas esenciales para manipulación de datos, codificación de variables, escalado, división de conjuntos, entrenamiento y evaluación de modelos de machine learning, junto con herramientas para interacción en el notebook.


```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import ipywidgets as widgets
from IPython.display import display, clear_output
```

## 🧹🔢 Limpieza y codificación de datos

Se eliminan los valores nulos y se codifican las variables categóricas con LabelEncoder, que transforma textos en números para que puedan ser interpretados por los algoritmos de machine learning.

```python
# Limpiar datos
df = df.dropna()  # Eliminamos filas con datos faltantes por seguridad

# Variables categóricas y numéricas
cat_cols = ['Job Title', 'Company', 'Location', 'Experience Level', 'Industry', 'Required Skills']
target_col = 'Salary'

# Codificamos las variables categóricas con LabelEncoder
#for col in cat_cols:
#    df[col] = LabelEncoder().fit_transform(df[col].astype(str))
```

## 💰📊 Clasificación del salario en categorías

Se crean tres categorías salariales (Inferior, Promedio y Superior) según rangos definidos, y se asignan a una nueva columna llamada `Salary Category`.



```python
# Clasificar los salarios en 3 categorías: Inferior, Promedio y Superior
def clasificar_salario(salario):
    if salario < 68000:
        return 'Salario Inferior'
    elif 68000 <= salario <= 123000:
        return 'Salario Promedio'
    else:
        return 'Salario Superior'

df['Salary Category'] = df[target_col].apply(clasificar_salario)

# Clasificamos los salarios en tres categorías utilizando los percentiles del resumen estadístico:
# - Menor al percentil 25 (68000): 'Salario Inferior'
# - Entre el percentil 25 y el 75 (68000 a 123000): 'Salario Promedio'
# - Mayor al percentil 75 (123000): 'Salario Superior'
# Esto nos permite segmentar los sueldos según su distribución en el conjunto de datos.
```

## 🎯 Definición de variables para el modelo

Se redefine la variable objetivo como `Salary Category` y se separan las variables independientes (features) y la variable dependiente (target) para el entrenamiento del modelo.

```python
# Ahora cambiamos el objetivo de 'Salary' a 'Salary Category'
target_col = 'Salary Category'

# Variables independientes (features) y la variable dependiente (target)
X = df.drop(columns=[target_col])
y = df[target_col]
```

## 📊 Codificación y estandarización de variables

Se codifican nuevamente las variables categóricas del conjunto `X` con `LabelEncoder`, y se estandarizan las variables numéricas usando `StandardScaler` para mejorar el rendimiento de los modelos.


```python
# Codificamos las variables categóricas
for col in cat_cols:
    X[col] = LabelEncoder().fit_transform(X[col].astype(str))

# Estandarizamos solo las variables numéricas (features)
num_cols = X.select_dtypes(include=[np.number]).columns.tolist()
scaler = StandardScaler()
X[num_cols] = scaler.fit_transform(X[num_cols])
```

## 🔀 Añadir ruido a los datos

Se define una función que agrega ruido aleatorio a las variables numéricas, con el fin de aumentar la complejidad del modelo y evitar sobreajuste. La semilla garantiza reproducibilidad del experimento.


```python
# Función para agregar ruido a los datos (más ruidoso)
def add_noise(data, noise_level=0.1):
    noisy_data = data.copy()
    for col in noisy_data.select_dtypes(include=['float64', 'int64']).columns:
        noise = np.random.normal(0, noise_level, noisy_data[col].shape)
        noisy_data[col] = noisy_data[col] + noise
    return noisy_data

np.random.seed(42)  # Semilla fija para ruido reproducible
# Agregar ruido a las características
X_noisy = add_noise(X, noise_level=0.1)  # Ruido aumentado
```

## 🧪 División del dataset en entrenamiento y prueba
Se divide el conjunto de datos en entrenamiento (80%) y prueba (20%) usando `train_test_split`, para evaluar el rendimiento de los modelos con datos no vistos.



```python
# División en train y test
X_train, X_test, y_train, y_test = train_test_split(X_noisy, y, test_size=0.2, random_state=42)
```

## 🤖 Entrenamiento y optimización de modelos de clasificación  
Se entrenan tres modelos de clasificación utilizando los datos de entrenamiento y se optimizan sus hiperparámetros con `GridSearchCV` para mejorar su desempeño:

- **Regresión Logística**  
- **Árbol de Decisión**  
- **Random Forest**  

GridSearchCV realiza una búsqueda exhaustiva en una rejilla de parámetros, usando validación cruzada para encontrar la mejor combinación de hiperparámetros y evitar sobreajuste.


```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV

# === Logistic Regression
param_grid_lr = {
    'C': [0.01, 0.1],
    'solver': ['liblinear', 'lbfgs']
}

grid_lr = GridSearchCV(LogisticRegression(max_iter=500, random_state=42), param_grid_lr, cv=2)
grid_lr.fit(X_train, y_train)
y_pred_lr = grid_lr.predict(X_test)
print("\nLogistic Regression best params:", grid_lr.best_params_)
print("Accuracy Logistic Regression:", grid_lr.score(X_test, y_test))


# === Decision Tree
param_grid_dt = {
    'max_depth': [2, 3],
    'min_samples_split': [4, 5]
}

grid_dt = GridSearchCV(DecisionTreeClassifier(random_state=42), param_grid_dt, cv=2)
grid_dt.fit(X_train, y_train)
y_pred_dt = grid_dt.predict(X_test)
print("\nDecision Tree best params:", grid_dt.best_params_)
print("Accuracy Decision Tree:", grid_dt.score(X_test, y_test))


# === Random Forest (rangos amplios para mejorar) ===
param_grid_rf = {
    'n_estimators': [40, 50],     # más árboles para mejorar estabilidad
    'max_depth': [5, 10, None],    # mayor profundidad para mayor complejidad
    'min_samples_split': [2, 3]          # pocos splits para flexibilidad
}

grid_rf = GridSearchCV(RandomForestClassifier(random_state=42), param_grid_rf, cv=5)
grid_rf.fit(X_train, y_train)
y_pred_rf = grid_rf.predict(X_test)
print("\nRandom Forest best params:", grid_rf.best_params_)
print("Accuracy Random Forest:", grid_rf.score(X_test, y_test))
```

## 📊 Evaluación del rendimiento de los modelos  
Se define una función para evaluar y mostrar las métricas clave de desempeño de cada modelo entrenado, tales como:

- Precisión (accuracy)  
- Matriz de confusión  
- Reporte de clasificación (precision, recall, f1-score)  

Estas métricas permiten comparar la efectividad y calidad de las predicciones de cada modelo.




```python
def print_metrics(y_true, y_pred, model_name):
    accuracy = accuracy_score(y_true, y_pred)
    print(f"{model_name} ")
    print(f"Accuracy: {accuracy:.4f}")
    print(confusion_matrix(y_true, y_pred))
    print(classification_report(y_true, y_pred))
    print("\n")

# Imprimir las métricas para cada modelo
print_metrics(y_test, y_pred_lr, "Regresión Logística")
print_metrics(y_test, y_pred_dt, "Árbol de Decisión")
print_metrics(y_test, y_pred_rf, "Random Forest")
```

## 📈 Análisis de la Matriz de Confusión  
Se visualizan las matrices de confusión de los modelos **Regresión Logística**, **Árbol de Decisión** y **Random Forest** usando Plotly, para evaluar su desempeño en la clasificación de las tres categorías salariales:  

- Salario Inferior  
- Salario Promedio  
- Salario Superior  

Esta visualización interactiva facilita la interpretación de los aciertos y errores de cada modelo.

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import pandas as pd
from sklearn.metrics import confusion_matrix

# Etiquetas de clase
labels = ["Salario Inferior", "Salario Promedio", "Salario Superior"]

# Crear la figura con 3 subgráficos horizontales
fig = make_subplots(
    rows=1, cols=3,
    subplot_titles=("Regresión Logística", "Árbol de Decisión", "Random Forest"),
    shared_yaxes=True
)

# Función para agregar cada matriz a los subplots
def add_confusion_to_subplot(fig, y_true, y_pred, col, scale='Reds'):
    cm = confusion_matrix(y_true, y_pred)
    fig.add_trace(
        go.Heatmap(
            z=cm,
            x=labels, # Predicción del modelo
            y=labels, # Clase real
            colorscale=scale,
            colorbar=dict(title="Count") if col == 3 else None,
            showscale=True if col == 3 else False,
            hovertemplate="Clase real: %{y}<br>Predicción del modelo: %{x}<br>Count: %{z}<extra></extra>"

        ),
        row=1, col=col
    )

# Agregar los 3 modelos
add_confusion_to_subplot(fig, y_test, y_pred_lr, col=1, scale='Blues')
add_confusion_to_subplot(fig, y_test, y_pred_dt, col=2, scale='Blues')
add_confusion_to_subplot(fig, y_test, y_pred_rf, col=3, scale='Blues')

# Ajustes de layout
fig.update_layout(
    title_text="Matrices de Confusión Comparadas por Modelo",
    height=500,
    width=1300
)

# Mostrar figura
fig.show()
```

## 📊 Comparación de Precisión entre Modelos  
Se comparan los modelos `'Regresión Logística'`, `'Árbol de Decisión'` y `'Random Forest'` en función de su precisión, permitiendo identificar cuál ofrece el mejor rendimiento para la clasificación de categorías salariales.  
La comparación se realiza mediante un gráfico de barras horizontales generado con Plotly para una visualización clara y efectiva.

```python
import plotly.graph_objects as go
from sklearn.metrics import accuracy_score, classification_report

# Definir una función para obtener precisión y otras métricas importantes
def get_metrics(y_true, y_pred):
    accuracy = accuracy_score(y_true, y_pred)
    report = classification_report(y_true, y_pred, output_dict=True)
    return accuracy, report['accuracy'], report['macro avg']['f1-score']

# Calcular métricas para cada modelo
lr_accuracy, lr_accuracy_detail, lr_f1 = get_metrics(y_test, y_pred_lr)
dt_accuracy, dt_accuracy_detail, dt_f1 = get_metrics(y_test, y_pred_dt)
rf_accuracy, rf_accuracy_detail, rf_f1 = get_metrics(y_test, y_pred_rf)

# Crear un diccionario con los resultados de precisión
models = {
    'Regresión Logística': lr_accuracy,
    'Árbol de Decisión': dt_accuracy,
    'Random Forest': rf_accuracy
}

# Colores para cada modelo
colors = {
    'Regresión Logística': '#1f77b4',  # azul
    'Árbol de Decisión': '#ff7f0e',    # naranja
    'Random Forest': '#2ca02c'          # verde
}

# Ordenar los modelos por precisión (de mayor a menor)
sorted_models = sorted(models.items(), key=lambda x: x[1], reverse=True)

# Preparar los datos ordenados
model_names = [model[0] for model in sorted_models]
accuracies = [model[1] for model in sorted_models]

# Crear gráfico de barras horizontal
fig = go.Figure(data=[
    go.Bar(
        y=[model_names[i]],
        x=[accuracies[i]],
        orientation='h',  # barras horizontales
        text=[f"{accuracies[i]:.4f}"],  # texto con precisión
        textposition='auto',
        textfont=dict(size=16),
        marker=dict(color=colors[model_names[i]])
    ) for i in range(len(model_names))
])

# Mejoras de estilo
fig.update_layout(
    title='Comparación de Modelos - Precisión',
    xaxis_title='Precisión',
    yaxis_title='Modelo',
    yaxis=dict(autorange='reversed'),  # para mostrar la barra de mayor a menor de arriba a abajo
    width=1300,
    height=500,
    showlegend=False,
)

fig.show()
```

## 📈 Reporte de Impacto Económico de Modelos de Clasificación 💰

Este módulo permite generar un reporte comparativo del impacto económico para los modelos de clasificación seleccionados:

**Random Forest**, **Árbol de Decisión** y **Regresión Logística**.

El reporte incluye una tabla detallada que muestra, por cada categoría salarial (Salario Inferior, Promedio y Superior):
- La cantidad de aciertos (verdaderos positivos)
- Errores de tipo falso positivo y falso negativo
- Beneficios obtenidos por acierto
- Penalizaciones por errores
- Impacto económico neto

La interfaz cuenta con dos botones:
- **Generar Reporte:** despliega la tabla con los datos anteriores.
- **Mostrar Gráfico:** presenta visualizaciones de barras que ilustran:
  - Impacto económico por categoría salarial en USD
  - Penalización total, aciertos totales e impacto neto total para el modelo seleccionado

Estas herramientas facilitan una comparación clara y práctica del desempeño económico de cada modelo.

```python
import ipywidgets as widgets
from IPython.display import display, clear_output, HTML
import pandas as pd
import numpy as np
from sklearn.metrics import confusion_matrix
import altair as alt

# Diccionario de modelos y sus predicciones
modelos_dict = {
    "Random Forest": {"y_pred": y_pred_rf, "modelo": grid_rf},
    "Árbol de Decisión": {"y_pred": y_pred_dt, "modelo": grid_dt},
    "Regresión Logística": {"y_pred": y_pred_lr, "modelo": grid_lr},
}

# Widgets
selector_modelo = widgets.RadioButtons(
    options=list(modelos_dict.keys()),
    description='Modelo:',
    style={'description_width': 'initial'}
)
boton_generar = widgets.Button(description="Generar Reporte", button_style="success")
boton_grafico = widgets.Button(description="Mostrar Gráfico", button_style="info")
salida = widgets.Output()
salida_grafico = widgets.Output()

df_impacto_global = None  # Variable global
modelo_actual = None      # Para guardar nombre del modelo seleccionado

# Tabla con impacto económico
def tabla_impacto_economico(y_true, y_pred):
    valores = {
        'Salario Inferior':    {'beneficio': 50,   'costo_fp': 200, 'costo_fn': 100},
        'Salario Promedio':    {'beneficio': 80,   'costo_fp': 250, 'costo_fn': 150},
        'Salario Superior':    {'beneficio': 100,  'costo_fp': 500, 'costo_fn': 200},
    }
    etiquetas = ['Salario Inferior', 'Salario Promedio', 'Salario Superior']
    cm = confusion_matrix(y_true, y_pred, labels=etiquetas)

    datos = []
    total_beneficio = 0
    total_penalizacion = 0

    for i, clase in enumerate(etiquetas):
        vp = cm[i, i]
        fn = cm[i, :].sum() - vp
        fp = cm[:, i].sum() - vp

        beneficio = vp * valores[clase]['beneficio']
        penalizacion = (fp * valores[clase]['costo_fp']) + (fn * valores[clase]['costo_fn'])
        impacto_neto = beneficio - penalizacion

        total_beneficio += beneficio
        total_penalizacion += penalizacion

        datos.append({
            "Clase": clase,
            "Aciertos (VP)": vp,
            "Errores FP": fp,
            "Errores FN": fn,
            "Beneficio por Acierto": beneficio,
            "Penalización": penalizacion,
            "Impacto Neto": impacto_neto
        })

    total_neto = total_beneficio - total_penalizacion
    datos.append({
        "Clase": "💰 TOTAL 💰",
        "Aciertos (VP)": "",
        "Errores FP": "",
        "Errores FN": "",
        "Beneficio por Acierto": total_beneficio,
        "Penalización": total_penalizacion,
        "Impacto Neto": total_neto
    })

    df = pd.DataFrame(datos)

    # Formatear columnas monetarias
    df_tabla = df.copy()
    for col in ['Beneficio por Acierto', 'Penalización', 'Impacto Neto']:
        df_tabla[col] = df_tabla[col].apply(lambda x: f"USD $ {x:,.0f}" if pd.notna(x) and x != "" else x)

    display(HTML(df_tabla.to_html(index=False)))
    return df

# Gráfico con Altair
def graficar_impacto(df, nombre_modelo):
    df_grafico = df[df["Clase"] != "💰 TOTAL 💰"].copy()

    orden_metricas = ['Penalización', 'Beneficio por Acierto', 'Impacto Neto']
    df_melt = df_grafico.melt(
        id_vars='Clase',
        value_vars=orden_metricas,
        var_name='Métrica',
        value_name='Valor'
    )

    colores = alt.Scale(
        domain=orden_metricas,
        range=['#e74c3c', '#27ae60', '#2980b9']
    )

    chart_clases = alt.Chart(df_melt).mark_bar().encode(
        x=alt.X('Clase:N', title='Clase', axis=alt.Axis(labelAngle=30)),
        y=alt.Y('Valor:Q', title='Monto - USD'),
        color=alt.Color('Métrica:N', scale=colores, title='Métrica'),
        xOffset='Métrica:N',
        tooltip=['Clase', 'Métrica', alt.Tooltip('Valor', format=',')]
    ).properties(
        width=600,
        height=400,
        title=f'Impacto Económico por Clase del Modelo: {nombre_modelo}'
    )

    # Texto encima de las barras
    text = alt.Chart(df_melt).mark_text(
        align='center',
        baseline='bottom',
        dy=-5,
        fontSize=12
    ).encode(
        x=alt.X('Clase:N', axis=alt.Axis(labelAngle=30)),
        y='Valor:Q',
        text=alt.Text('Valor:Q', format=','),
        detail='Métrica:N',
        xOffset='Métrica:N'
    )

    chart_clases = (chart_clases + text)

    # Gráfico resumen total
    df_total = df[df["Clase"] == "💰 TOTAL 💰"].copy()
    resumen_total = pd.DataFrame({
        "Métrica": orden_metricas,
        "Valor": [
            df_total["Penalización"].values[0],
            df_total["Beneficio por Acierto"].values[0],
            df_total["Impacto Neto"].values[0]
        ]
    })

    chart_total = alt.Chart(resumen_total).mark_bar().encode(
        x=alt.X('Métrica:N', title='Métrica', axis=alt.Axis(labelAngle=0)),
        y=alt.Y('Valor:Q', title='Monto Total - USD'),
        color=alt.Color('Métrica:N', legend=None, scale=colores),
        tooltip=[alt.Tooltip('Métrica'), alt.Tooltip('Valor', format=',')]
    ).properties(
        width=600,
        height=400,
        title=f'💼 Resumen Económico Total del Modelo: {nombre_modelo}'
    )

    # Texto en barras del resumen total
    text_total = alt.Chart(resumen_total).mark_text(
        align='center',
        baseline='bottom',
        dy=-5,
        fontSize=12
    ).encode(
        x='Métrica:N',
        y='Valor:Q',
        text=alt.Text('Valor:Q', format=',')
    )

    chart_total = (chart_total + text_total)

    combined_chart = alt.hconcat(chart_clases, chart_total).resolve_scale(y='independent')

    # Guardar los gráficos como variables globales para futura exportación
    global altair_chart_clases, altair_chart_total
    altair_chart_clases = chart_clases
    altair_chart_total = chart_total

    with salida_grafico:
        clear_output()
        display(combined_chart)

# Acción del botón Generar Reporte
def generar_reporte(b):
    global df_impacto_global, modelo_actual
    with salida:
        clear_output()
        modelo_actual = selector_modelo.value
        y_pred = modelos_dict[modelo_actual]["y_pred"]
        df_impacto_global = tabla_impacto_economico(y_test, y_pred)
    # Limpiar también la salida del gráfico al generar un reporte nuevo
    with salida_grafico:
        clear_output()

# Acción del botón Mostrar Gráfico
def mostrar_grafico(b):
    if df_impacto_global is not None and modelo_actual is not None:
        graficar_impacto(df_impacto_global, modelo_actual)
    else:
        with salida_grafico:
            clear_output()
            print("Primero genera el reporte.")

# Conectar botones
boton_generar.on_click(generar_reporte)
boton_grafico.on_click(mostrar_grafico)

# Mostrar interfaz
display(selector_modelo, widgets.HBox([boton_generar, boton_grafico]), salida, salida_grafico)
```

## 💾 Exportar Reporte Económico en PDF  
Este bloque genera un botón llamado **"ExportarReporte.PDF"** que permite descargar un reporte completo del impacto económico del modelo seleccionado.  

El reporte incluye:  
- Un título descriptivo del análisis económico.  
- La tabla de impacto económico con detalles por clase (aciertos, errores, beneficios y penalizaciones).  
- Los gráficos de barras que muestran el impacto económico por categoría salarial (Salario Inferior, Promedio y Superior).  
- Un resumen económico total del modelo seleccionado con visualizaciones comparativas.  

Esta funcionalidad facilita compartir y analizar los resultados económicos del modelo de manera clara y profesional.

```python
import ipywidgets as widgets
from IPython.display import display, clear_output, FileLink, IFrame
import os
import re

# función exportadora
from reportlab.lib.pagesizes import letter, landscape
from reportlab.platypus import SimpleDocTemplate, Table, TableStyle, Paragraph, Image, Spacer
from reportlab.lib import colors
from reportlab.lib.styles import getSampleStyleSheet

def exportar_reporte_pdf_reportlab(nombre_modelo, df_tabla, chart1, chart2, ruta_archivo):
    chart1_path = "chart1.png"
    chart2_path = "chart2.png"
    chart1.save(chart1_path, scale_factor=2)
    chart2.save(chart2_path, scale_factor=2)

    doc = SimpleDocTemplate(ruta_archivo, pagesize=landscape(letter),
                            rightMargin=30, leftMargin=30, topMargin=30, bottomMargin=30)
    elements = []

    styles = getSampleStyleSheet()
    style_title = styles['Title']
    style_normal = styles['Normal']

    elements.append(Paragraph(f"Reporte del Modelo: {nombre_modelo}", style_title))
    elements.append(Spacer(1, 12))

    data = [df_tabla.columns.tolist()]

    def formatear_valor(valor, col):
        if col in ['Beneficio por Acierto', 'Penalización', 'Impacto Neto']:
            try:
                num = float(str(valor).replace(',', '').replace('$', ''))
                return f"${num:,.0f}"
            except:
                return str(valor)
        else:
            return str(valor)

    for _, row in df_tabla.iterrows():
        fila = [formatear_valor(row[col], col) for col in df_tabla.columns]
        data.append(fila)

    num_cols = len(df_tabla.columns)
    page_width = landscape(letter)[0] - 60
    col_width = page_width / num_cols
    col_widths = [col_width] * num_cols

    table = Table(data, colWidths=col_widths, hAlign='CENTER')

    style = TableStyle([
        ('BACKGROUND', (0,0), (-1,0), colors.HexColor('#d3d3d3')),
        ('TEXTCOLOR', (0,0), (-1,0), colors.black),
        ('ALIGN', (0,0), (0,-1), 'LEFT'),
        ('ALIGN', (1,0), (-1,-1), 'RIGHT'),
        ('FONTNAME', (0,0), (-1,0), 'Helvetica-Bold'),
        ('FONTSIZE', (0,0), (-1,-1), 8),
        ('BOTTOMPADDING', (0,0), (-1,0), 8),
        ('GRID', (0,0), (-1,-1), 0.5, colors.black),
    ])

    table.setStyle(style)

    elements.append(table)
    elements.append(Spacer(1, 20))

    img1 = Image(chart1_path, width=450, height=250)
    img2 = Image(chart2_path, width=450, height=250)
    elements.append(img1)
    elements.append(Spacer(1, 12))
    elements.append(img2)

    doc.build(elements)

    os.remove(chart1_path)
    os.remove(chart2_path)

    print(f"✅ PDF generado con reportlab: {ruta_archivo}")


# --- Variables de ejemplo ---
# modelo_actual = "Modelo X"
# df_impacto_global = dataframe
# altair_chart_clases = grafico1
# altair_chart_total = grafico2
ruta_pdf = f"reporte_modelo_impacto_Economico.pdf"

boton_exportar = widgets.Button(description="ExportarReporte.PDF", button_style='success')
output = widgets.Output()

def on_exportar_clicked(b):
    with output:
        clear_output()
        exportar_reporte_pdf_reportlab(modelo_actual, df_impacto_global, altair_chart_clases, altair_chart_total, ruta_pdf)
        if os.path.exists(ruta_pdf):
            display(widgets.HTML(f'<b>Archivo generado:</b> {ruta_pdf}'))
            display(FileLink(ruta_pdf))
            display(IFrame(src=ruta_pdf, width=1000, height=400))

boton_exportar.on_click(on_exportar_clicked)
display(boton_exportar, output)
```
